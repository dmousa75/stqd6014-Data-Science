```python
!pip install squarify
import squarify
import matplotlib.pyplot as plt
import pandas as pd

# Group data by 'NamaDUN' and sum votes for each party
grouped_data = pru15_df.groupby('NamaDUN')[['BN', 'PH', 'PN']].sum().reset_index()

# Calculate total votes for each 'NamaDUN'
grouped_data['Total Votes'] = grouped_data['BN'] + grouped_data['PH'] + grouped_data['PN']

# Create a list of labels for each rectangle
labels = [f"{name}\nTotal Votes: {votes}" for name, votes in zip(grouped_data['NamaDUN'], grouped_data['Total Votes'])]

# Create the treemap
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
ax = squarify.plot(sizes=grouped_data['Total Votes'], label=labels, alpha=.8)

# --- Adding Legend for Party Percentages ---
# Calculate percentages for each party
total_votes = grouped_data['Total Votes'].sum()
bn_percent = (grouped_data['BN'].sum() / total_votes) * 100
ph_percent = (grouped_data['PH'].sum() / total_votes) * 100
pn_percent = (grouped_data['PN'].sum() / total_votes) * 100

# Create legend handles and labels
handles = [
    plt.Rectangle((0, 0), 1, 1, color='blue'),  # Assuming BN is blue
    plt.Rectangle((0, 0), 1, 1, color='red'),   # Assuming PH is red
    plt.Rectangle((0, 0), 1, 1, color='green')  # Assuming PN is green
]
labels = [
    f"BN: {bn_percent:.1f}%",
    f"PH: {ph_percent:.1f}%",
    f"PN: {pn_percent:.1f}%"
]

# Add legend to the plot
plt.legend(handles, labels, title="Party Percentages", loc="upper right")
# --- End of Legend Addition ---

plt.axis('off')  # Hide axes
plt.title('Total Votes Density among NamaDUN (Treemap)')
plt.show()
```
