## Remove certain rows

```python
# Assuming the row we want to remove
row_label_remove = 'AYER KUNING ALL', 'JUMLAH PENGUNDI', 'JUMLAH AHLI UMNO'
```

```python
# Drop the row
df = df[df['PDM'] != row_label_remove]
```

```python
# Check gain the dataset
df['PDM']
```

## Groupby Categories by 'PDM'

### All categories

```python
# Groupby 'PDM'
df_1 = df.groupby('PDM').sum()
df_1.head()
```

### PRN Votes & Ahli UMNO by PDM

#### Convert to numerics

```python
df_1
```

```python
print(df_1.columns)
```

```python
columns_to_convert = ['UNDI_DUN_BN', 'UNDI_DUN_PH', 'UNDI_DUN_PN', 'DEWASA_umno', 'WANITA_umno','PEMUDA_umno ', 'PUTERI_umno']

for column in columns_to_convert:
    df_1[column] = pd.to_numeric(df_1[column], errors='coerce').astype(object)

```

```python
df_1.info()
```

```python
df_2 = df.groupby('PDM').sum()[['UNDI_DUN_BN','UNDI_DUN_PH', 'UNDI_DUN_PN', 'DEWASA_umno','WANITA_umno', 'PEMUDA_umno ', 'PUTERI_umno']]
df_2.head()
```

## Visualizations

### DUN Votes & Ahli UMNO

```python
df_2.plot(kind='bar')
plt.show()
```

### Change to Long Format

```python
# Reshape the DataFrame to long format
df_long = pd.melt(df_2.reset_index(),
                 id_vars=['PDM'],
                 value_vars=['UNDI_DUN_BN', 'UNDI_DUN_PH', 'UNDI_DUN_PN', 'DEWASA_umno', 'WANITA_umno', 'PEMUDA_umno ', 'PUTERI_umno'],
                 var_name='Category',
                 value_name='Value')

# Display the reshaped DataFrame
print(df_long.head())
```

### Barplot Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='PDM', y='Value', hue='Category', data=df_long)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.title('Comparison of Categories by PDM')
plt.tight_layout()
plt.show()
```

#### Add Sum Values to each Bar

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a bar plot using seaborn
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='PDM', y='Value', hue='Category', data=df_long)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.title('Comparison of Categories by PDM', size= 15, fontweight= 'bold')

# Add sum values on each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 10),
                textcoords = 'offset points')

# Calculate total sums for each category
total_sums = df_long.groupby('Category')['Value'].sum()

# Update legend with total sums
handles, labels = ax.get_legend_handles_labels()
for i, label in enumerate(labels):
    labels[i] = f"{label} (Total: {total_sums[label]:.0f})"
ax.legend(handles, labels, title="Category (Total Sum)")

plt.tight_layout()
plt.show()
```

### Pie Chart


#### Pie Chart 1 & Setting UP

```python
# Data for the pie charts
undi_dun_data = df_2[['UNDI_DUN_BN', 'UNDI_DUN_PH', 'UNDI_DUN_PN']].sum()
ahli_umno_data = df_2[['DEWASA_umno', 'WANITA_umno', 'PEMUDA_umno ', 'PUTERI_umno']].sum()

# Labels for the segments
undi_dun_labels = ['UNDI_DUN_BN', 'UNDI_DUN_PH', 'UNDI_DUN_PN']
ahli_umno_labels = ['DEWASA_umno', 'WANITA_umno', 'PEMUDA_umno', 'PUTERI_umno']

# --- Pie Chart 1: UNDI_DUN ---
plt.figure(figsize=(6, 6))  # Adjust figure size as needed
plt.pie(undi_dun_data, labels=undi_dun_labels, autopct='%1.1f%%', startangle=90)
plt.title('Undi DUN Distribution')
```

#### Pie Chart 2

```python
# --- Pie Chart 2: Ahli UMNO ---
plt.figure(figsize=(6, 6))  # Adjust figure size as needed
plt.pie(ahli_umno_data, labels=ahli_umno_labels, autopct='%1.1f%%', startangle=90)
plt.title('Ahli UMNO Distribution')
```
