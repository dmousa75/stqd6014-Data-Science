```python
import pandas as pd
import numpy as np
```

```python
#filling missing data

#create mock data
np.random.seed(1312)
df = pd.DataFrame(np.random.standard_normal((7,3)))
df
```

```python
#randomly add missing data
#iloc - integer location based indexing
df.iloc[:4,1] = np.nan #all columns before row 4 column 1
df.iloc[:2, 2] = np.nan #all columns before row 2 column 2
df

```

```python
#filling the NaN with a constant
#Sentinel data -> placeholder
#changes happens only by display, it wouldnt change the asal file
df_edit = df.fillna(-999)
df_edit
```

```python
#df asal dont change!
df
```

```python
[] # used to call fillna with a different fill for each columns
#change by display only

df.fillna({1:0.5, 2: 0})#fill all "NaN Values' at column 1 & column 2
```

```python
#but all we did above didnt affect the original df
df
```

```python
# but if we want to change permanent eg at (2,0) to 0.37, it will change
# permanently ( at row 2 column 0)
df.iloc[2,0] = 0.37
df
```

```python
np.random.seed(1234)
df = pd.DataFrame(np.random.standard_normal((6,3)))
df
```

```python
#iloc- integer index
df.iloc[2:, 1] = np.nan # at row 2 and after, column 1
df.iloc[4: , 2] = np.nan # at row 4 and after, column 2
df
```

```python
#'ffill' -> forward fill
#uses the previous valid (non missing) value to replace the NaN( value row sebelah atas, sebelum NaN)
df.ffill() # kalau nak limit brp row yg berlaku, masukkan dlm parenthesis tu jum brpr nak eg 'df.ffill(2)'
```

```python
# Mock data
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'pastrami', 'corned beef', 'bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
# Mapping of each distinct animal type
# This is a library of food to animal
meat_to_animal = {
    "bacon": "pig",
    "pulled pork": "pig",
    "pastrami": "cow",
    "corned beef": "cow",
    "honey ham": "pig",
    "nova lox": "salmon"
}
meat_to_animal
```

```python
# Mapping method - so its will recip values in library of food to animal, make
#new colum named 'animal'
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
data
```
