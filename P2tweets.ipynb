# Introduction of Dataset "tweets"

About Dataset

The "Indian Politics Tweets & Reactions" dataset contains a collection of public tweets and their associated reactions related to Indian politics. The dataset provides insights into public opinion and sentiment on various political issues and events in India, as expressed on Twitter.

The dataset includes a diverse range of tweets from users across the country, covering topics such as political campaigns, government policies, elections, political parties, leaders, and controversies.

The reactions to each tweet are also included, such as likes, retweets, and comments, allowing for a comprehensive analysis of the engagement and impact of each tweet.

Researchers and analysts can use this dataset to study the sentiment and trends of public opinion on Indian politics and to gain insights into the effectiveness of political campaigns and messaging on social media.

# Data Cleaning

```python
#check 'NaN' values
tweets.isnull()
```

```python
#find 'NaN' at which variables
tweets.isnull().any()
```

```python
#filter how much the 'NaN' at variables
tweets.isnull().sum()
```

```python
print(tweets.shape)
```

```python
print(tweets)
```

```python
#filters out which columns have the missing data
nan_columns = [c for c in tweets.columns if tweets[c].isnull().any()]
print(nan_columns)
```

```python
#essentially filter out the rows with missing data
nan_rows = tweets[tweets.isnull().any(axis=1)]
print(nan_rows)
```

```python
#drop the missing values any rows in dataframe
#ang check again if there any missing values
tweets = tweets.dropna()
tweets.isnull().sum()
```

```python
#display the sructure of DataFrame
print(tweets.info())
```

```python
#print columns names
print(tweets.columns)
```

```python
#count the number of tweets, retweets and replies
num_tweets = tweets['Tweet'].count()
num_retweets = tweets['Retweets'].count()
num_likes = tweets['Likes'].count()

print("Number of tweets:", num_tweets)
print("Number of retweets:", num_retweets)
print("Number of likes:", num_likes)
```
